### What is named entity?

Classifying the various parts of text in categories.

These list of categories is fixed, Some are given below.


> nltk's NE feature can be useful but it has high error rate and high false positive.
>
> Sometimes we need to use Nouns and sometimes we need to use Named Entity.

```python
import nltk
from nltk.corpus import state_union # dataset of presidents state of union from past 70 years
from nltk.tokenize import PunktSentenceTokenizer # unsupervised ml sentece_tokenizer, it comes pre trained but we can retrain it for ourselves.
```

```python
train_text = state_union.raw("2005-GWBush.txt")
sample_text = state_union.raw("2006-GWBush.txt")
```

```python
custom_sent_tokenize = PunktSentenceTokenizer(train_text) # train the model
tokenized = custom_sent_tokenize.tokenize(sample_text)
```

```python
def process_content():
    try:
        for sent in tokenized[5:]:
            words = nltk.word_tokenize(sent)
            tagged = nltk.pos_tag(words)
            
            namedEnt = nltk.ne_chunk(tagged, binary=True)
            namedEnt.draw()
            
    except Exception as e:
        print(str(e))
```

```python
process_content()
```
